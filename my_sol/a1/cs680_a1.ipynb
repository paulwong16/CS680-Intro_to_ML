```
import pandas as pd
import numpy as np
from sklearn.utils import shuffle

# read from google drive, run on colab
# spambase_X = pd.read_csv('/content/drive/My Drive/Colab Notebooks/spambase_X.csv')
# spambase_y = pd.read_csv('/content/drive/My Drive/Colab Notebooks/spambase_y.csv')

# read from loacl
spambase_X = pd.read_csv('./spambase_X.csv')
spambase_y = pd.read_csv('./spambase_y.csv')

spambase_X = spambase_X.drop(spambase_X.columns[0], axis=1)

X = spambase_X.values.T
y = spambase_y.values
```

```
def perceptron(X,y,w,b,max_pass):
  n = X.shape[0]
  mistake = [0 for i in range(max_pass)]
  for t in range(max_pass):
    # X_train, y_train = shuffle(X, y)
    X_train, y_train = X, y
    for i in range(n):
      if (y_train[i]*(w.T.dot(X_train[i]) + b) <= 0):
        w = w + y_train[i]*X_train[i]
        b = b + y_train[i]
        mistake[t] = mistake[t] + 1

  return mistake
```

```
d = X.shape[1]
n = X.shape[0]
w = np.zeros(d)
b = 0

mistake = perceptron(X,y,w,b,500)
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(mistake)
```
